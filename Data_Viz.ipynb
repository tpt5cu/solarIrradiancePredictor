```python
#IMPORTS
import pandas as pd
import os
import seaborn as sns
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
def meta_visualization(data, x='GHI', y='DHI', z='Cloud Cover'):
    sns.lmplot(x, y, data=data, hue=z, fit_reg=False)
    plt.show()
    return
```

```python
def segment_visualization(data, x='GHI', y='DHI', z='Cloud Cover'):
    sns.lmplot(x, y, data=data, hue=z, row=z, fit_reg=True)
    plt.show()
    return
```

```python
def show_histograms(df):
    df['DHI'].hist()
    df['GHI'].hist()
    df['Cloud Cover'].hist()
```

```python
def read_dataframe(filepath):
    df = df = pd.read_csv(filepath)
    df = df[['DHI', 'DNI', 'GHI','Cloud Cover', 'timestamps']]
    print(df)
    return df
```
